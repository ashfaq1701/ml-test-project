```python
import pandas as pd

df = pd.read_csv('data/test.csv')
sentences = df['text'].tolist()
selected_sentences = sentences[:5]
selected_sentences
```

# Predict Using Linear Model

```python
import joblib
import numpy as np

model = joblib.load('results/saved_models/linear_classifier/model.joblib')

scores = model.decision_function(selected_sentences)
pred_indices = np.argmax(scores, axis=1)
labels = model.classes_[pred_indices]

for text, label, score_row in zip(selected_sentences, labels, scores):
    top_score = score_row[np.argmax(score_row)]
    print(f"{label} (score={top_score:.4f})")
```

# Predict Using Distilbert Model

```python
from pathlib import Path
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from distilbert_classifier.config import SAVED_MODELS_DIR

model_path = Path(SAVED_MODELS_DIR)  # results/saved_models/distilbert
device = 0 if torch.cuda.is_available() else -1  # GPU if available, else CPU

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

clf = pipeline("text-classification", model=model, tokenizer=tokenizer, device=device)

predictions = clf(selected_sentences)

for text, pred in zip(sentences, predictions):
    print(f"{pred['label']} (score={pred['score']:.4f})")
```
